```python
pca = prince.PCA(
    n_components=2,
    n_iter=10,
    rescale_with_mean=False,
    rescale_with_std=True,
    copy=True,
    check_input=True,
    engine='sklearn',
    random_state=42
)
pca = pca.fit(X)
```
