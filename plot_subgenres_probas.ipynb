```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_probs = pd.read_csv(r'PROBAS_PREDS_SUBGENRES.csv')
#df_probs = pd.read_csv(r'output_svm/output_GKF_canonizer_1850-1900_novel_scale.csv')
```

```python
df_probs.rename(columns={"Unnamed: 0": "index"}, inplace=True)
df_probs.set_index("index", inplace = True)
```

```python
def dated(df):
    df_res = df.copy()
    list_date = []
    for elem in df_res.index:
        index, date = shorted_str(elem)
        list_date.append(date)
    df_res['date'] = list_date
    return df_res
```

```python
df_probs_dated = dated(df_probs)
```

```python
sns.set_palette('bright')
sns.set_context("talk")
sns.set_style("darkgrid")
```

### AVENTURES

```python
plt.figure(figsize=(9,6))
ax = plt.subplot()
ax.set_title('Level of confidence of our model to assign aventures subgenre')

ax = sns.regplot(data=df_probs_dated, 
                 x='date', 
                 y='proba_aventures', 
                 color='blue', 
                 order=3, 
                 scatter=False, 
                 ax=ax)

ax.set_xlabel('Dates')
ax.set_ylabel('Probability assigned for aventures label')

plt.show()
```
